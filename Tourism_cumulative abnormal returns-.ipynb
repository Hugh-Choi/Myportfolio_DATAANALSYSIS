This appendix contains the relevant code for calculating Beta

Beta = Covariance (stock returns, market return) /Variance (market returns) 

```python
AA_beta=AA_Data.cov().iloc[0][1]/AA_Data.cov().iloc[0][0]
HT_beta=HT_Data.cov().iloc[0][1]/HT_Data.cov().iloc[0][0]
Hj_beta=Hj_Data.cov().iloc[0][1]/Hj_Data.cov().iloc[0][0]
HS_beta=HS_Data.cov().iloc[0][1]/HS_Data.cov().iloc[0][0]
LTD_beta=LTD_Data.cov().iloc[0][1]/LTD_Data.cov().iloc[0][0]
MN_beta=MN_Data.cov().iloc[0][1]/MN_Data.cov().iloc[0][0]
Pa_beta=Pa_Data.cov().iloc[0][1]/Pa_Data.cov().iloc[0][0]
RT_beta=RT_Data.cov().iloc[0][1]/RT_Data.cov().iloc[0][0]
Sj_beta=Sj_Data.cov().iloc[0][1]/Sj_Data.cov().iloc[0][0]
SS_beta=SS_Data.cov().iloc[0][1]/SS_Data.cov().iloc[0][0]
VGT_beta=VGT_Data.cov().iloc[0][1]/VGT_Data.cov().iloc[0][0]
KA_beta=KA_Data.cov().iloc[0][1]/KA_Data.cov().iloc[0][0]


Tourism_betas = pd.DataFrame({
                        "Asiana": str(AA_beta),
                        "Hanatour": str(HT_beta),
                        "Hanjin": str(Hj_beta),
                        "Hotel Silla": str(HS_beta),    
                        "Lotte tour development": str(LTD_beta),
                        "Modotour network": str(MN_beta),
                        "Paradise": str(Pa_beta),
                        "Redcap Tour": str(RT_beta),
                        "Sejoong": str(Sj_beta),
                        "Sinsegae": str(SS_beta),
                        "Verygoodtour": str(VGT_beta),
                        "Korea airline": str(KA_beta),
},index=["Tourism Beta"])

Tourism_betas.head()
```

```python
# Clean the data

AA_test_ret=AA_test.pct_change()
AA_test_ret=pd.DataFrame(AA_test_ret)
AA_test_ret=AA_test_ret.dropna() # here I dropped the NA terms

HT_test_ret=HT_test.pct_change()
HT_test_ret=pd.DataFrame(HT_test_ret)
HT_test_ret=HT_test_ret.dropna() # here I dropped the NA terms

Hj_test_ret=Hj_test.pct_change()
Hj_test_ret=pd.DataFrame(Hj_test_ret)
Hj_test_ret=Hj_test_ret.dropna() # here I dropped the NA terms

HS_test_ret=HS_test.pct_change()
HS_test_ret=pd.DataFrame(HS_test_ret)
HS_test_ret=HS_test_ret.dropna() # here I dropped the NA terms

LTD_test_ret=LTD_test.pct_change()
LTD_test_ret=pd.DataFrame(LTD_test_ret)
LTD_test_ret=LTD_test_ret.dropna() # here I dropped the NA terms

MN_test_ret=MN_test.pct_change()
MN_test_ret=pd.DataFrame(MN_test_ret)
MN_test_ret=MN_test_ret.dropna() # here I dropped the NA terms
```

```python
Pa_test_ret=Pa_test.pct_change()
Pa_test_ret=pd.DataFrame(Pa_test_ret)
Pa_test_ret=Pa_test_ret.dropna() # here I dropped the NA terms

RT_test_ret=RT_test.pct_change()
RT_test_ret=pd.DataFrame(RT_test_ret)
RT_test_ret=RT_test_ret.dropna() # here I dropped the NA terms

Sj_test_ret=HT_test.pct_change()
Sj_test_ret=pd.DataFrame(Sj_test_ret)
Sj_test_ret=HT_test_ret.dropna() # here I dropped the NA terms

SS_test_ret=SS_test.pct_change()
SS_test_ret=pd.DataFrame(SS_test_ret)
SS_test_ret=SS_test_ret.dropna() # here I dropped the NA terms

VGT_test_ret=VGT_test.pct_change()
VGT_test_ret=pd.DataFrame(VGT_test_ret)
VGT_test_ret=VGT_test_ret.dropna() # here I dropped the NA terms

KA_test_ret=KA_test.pct_change()
KA_test_ret=pd.DataFrame(KA_test_ret)
KA_test_ret=KA_test_ret.dropna() # here I dropped the NA terms

KS_test_ret=KS_test.pct_change()
KS_test_ret=pd.DataFrame(KS_test_ret)
KS_test_ret=KS_test_ret.dropna() # here I dropped the NA terms
```

```python
# First calculate abnormal returns
# Expected stock A’s return = Beta for stock A *(Market return - risk free rate) + risk free rate
# Abnormal Return = Actual return for the stock - Expected return for the stock

AA_Abn_ret = AA_test_ret-AA_beta*(KS_test_ret-(0.015/365))+0.015/365
HT_Abn_ret = HT_test_ret-HT_beta*(KS_test_ret-(0.015/365))+0.015/365
Hj_Abn_ret = Hj_test_ret-Hj_beta*(KS_test_ret-(0.015/365))+0.015/365
HS_Abn_ret = HS_test_ret-HS_beta*(KS_test_ret-(0.015/365))+0.015/365
LTD_Abn_ret = LTD_test_ret-LTD_beta*(KS_test_ret-(0.015/365))+0.015/365
MN_Abn_ret = MN_test_ret-MN_beta*(KS_test_ret-(0.015/365))+0.015/365
Pa_Abn_ret = Pa_test_ret-Pa_beta*(KS_test_ret-(0.015/365))+0.015/365
RT_Abn_ret = RT_test_ret-RT_beta*(KS_test_ret-(0.015/365))+0.015/365
Sj_Abn_ret = Sj_test_ret-Sj_beta*(KS_test_ret-(0.015/365))+0.015/365
SS_Abn_ret = SS_test_ret-SS_beta*(KS_test_ret-(0.015/365))+0.015/365
VGT_Abn_ret = VGT_test_ret-VGT_beta*(KS_test_ret-(0.015/365))+0.015/365
KA_Abn_ret = KA_test_ret-KA_beta*(KS_test_ret-(0.015/365))+0.015/365 
```

```python
Tourism_Abn_return = pd.DataFrame({
                        "Asiana": AA_Abn_ret["Adj Close"],
                        "Hanatour": HT_Abn_ret["Adj Close"],
                        "Hanjin": Hj_Abn_ret["Adj Close"],
                        "Hotel Silla": HS_Abn_ret["Adj Close"],    
                        "Lotte tour development": LTD_Abn_ret["Adj Close"],
                        "Modotour network": MN_Abn_ret["Adj Close"],
                        "Paradise": Pa_Abn_ret["Adj Close"],
                        "Redcap Tour": RT_Abn_ret["Adj Close"],
                        "Sejoong": Sj_Abn_ret["Adj Close"],
                        "Sinsegae": SS_Abn_ret["Adj Close"],
                        "Verygoodtour": VGT_Abn_ret["Adj Close"],
                        "Korea airline": KA_Abn_ret["Adj Close"],
})
 
Tourism_Abn_return.tail(22)
```

This appendix contains the relevant code for Cumulative Abnormal return

Cumulative abnormal return (CAR) is the sum of abnormal returns over a period, 30 day after the event date

```python
# note here the import thing to measure impact is CUMULATIVE ABNORMAL RETURN, NOT ABNORMAL RETURN, next we calculate CUMULATIVE ABNORMAL RETURN
#Cumulative Abnormal Return is the sum of abnormal returns over a period (several days)

AA_CAbn_ret=AA_Abn_ret.cumsum()
HT_CAbn_ret=HT_Abn_ret.cumsum()
Hj_CAbn_ret=Hj_Abn_ret.cumsum()
HS_CAbn_ret=HS_Abn_ret.cumsum()
LTD_CAbn_ret=LTD_Abn_ret.cumsum()
MN_CAbn_ret=MN_Abn_ret.cumsum()
Pa_CAbn_ret=Pa_Abn_ret.cumsum()
RT_CAbn_ret=RT_Abn_ret.cumsum()
Sj_CAbn_ret=Sj_Abn_ret.cumsum()
SS_CAbn_ret=SS_Abn_ret.cumsum()
VGT_CAbn_ret=VGT_Abn_ret.cumsum()
KA_CAbn_ret=KA_Abn_ret.cumsum()

Tourism_CAbn_return = pd.DataFrame({
                        "Asiana": AA_CAbn_ret["Adj Close"],
                        "Hanatour": HT_CAbn_ret["Adj Close"],
                        "Hanjin": Hj_CAbn_ret["Adj Close"],
                        "Hotel Silla": HS_CAbn_ret["Adj Close"],    
                        "Lotte tour development": LTD_CAbn_ret["Adj Close"],
                        "Modotour network": MN_CAbn_ret["Adj Close"],
                        "Paradise": Pa_CAbn_ret["Adj Close"],
                        "Redcap Tour": RT_CAbn_ret["Adj Close"],
                        "Sejoong": Sj_CAbn_ret["Adj Close"],
                        "Sinsegae": SS_CAbn_ret["Adj Close"],
                        "Verygoodtour": VGT_CAbn_ret["Adj Close"],
                        "Korea airline": KA_CAbn_ret["Adj Close"],
})
Tourism_CAbn_return.tail(1)
```
