```python
Ko_train=pdr.get_data_yahoo('024720.KS','2015-07-08','2016-07-08')['Adj Close']
LG_train=pdr.get_data_yahoo('051900.KS','2015-07-08','2016-07-08')['Adj Close']
Ne_train=pdr.get_data_yahoo('092730.KQ','2015-07-08','2016-07-08')['Adj Close']

Ko_train_ret=Ko_train.pct_change()
Ko_train_ret=pd.DataFrame(Ko_train_ret)
Ko_train_ret=Ko_train_ret.dropna() # here I dropped the NA terms

LG_train_ret=LG_train.pct_change()
LG_train_ret=pd.DataFrame(LG_train_ret)
LG_train_ret=LG_train_ret.dropna() # here I dropped the NA terms

Ne_train_ret=Ne_train.pct_change()
Ne_train_ret=pd.DataFrame(Ne_train_ret)
Ne_train_ret=Ne_train_ret.dropna() # here I dropped the NA terms
```

```python
KS_train=pdr.get_data_yahoo('^KS11','2015-07-08','2016-07-08')['Adj Close']

KS_train_ret=KS_train.pct_change()
KS_train_ret=pd.DataFrame(KS_train_ret)
KS_train_ret=KS_train_ret.dropna() # here I dropped the NA terms
```

```python
Ab_Data=pd.merge(KS_train_ret,Ab_train_ret,left_index=True,right_index=True)
AC_Data=pd.merge(KS_train_ret,AC_train_ret,left_index=True,right_index=True)
Co_Data=pd.merge(KS_train_ret,Co_train_ret,left_index=True,right_index=True) 
COS_Data=pd.merge(KS_train_ret,COS_train_ret,left_index=True,right_index=True) 
HCM_Data=pd.merge(KS_train_ret,HCM_train_ret,left_index=True,right_index=True) 
HC_Data=pd.merge(KS_train_ret,HC_train_ret,left_index=True,right_index=True) 
IH_Data=pd.merge(KS_train_ret,IH_train_ret,left_index=True,right_index=True) 
Ja_Data=pd.merge(KS_train_ret,Ja_train_ret,left_index=True,right_index=True) 
TO_Data=pd.merge(KS_train_ret,TO_train_ret,left_index=True,right_index=True) 
Ko_Data=pd.merge(KS_train_ret,Ko_train_ret,left_index=True,right_index=True) 
LG_Data=pd.merge(KS_train_ret,LG_train_ret,left_index=True,right_index=True) 
Ne_Data=pd.merge(KS_train_ret,Ne_train_ret,left_index=True,right_index=True) 


# It is usually safer to first merge two time-series using pd.merge(), it will make sure the dates between two series are matched

Ab_Data.columns=['KS','Ab']
AC_Data.columns=['KS','AC']
Co_Data.columns=['KS','Co']
COS_Data.columns=['KS','COS']
HCM_Data.columns=['KS','HCM']
HC_Data.columns=['KS','HC']
IH_Data.columns=['KS','IH']
Ja_Data.columns=['KS','Ja']
TO_Data.columns=['KS','TO']
Ko_Data.columns=['KS','Ko']
LG_Data.columns=['KS','LG']
Ne_Data.columns=['KS','Ne']
```

```python
# To calculate Beta
#Beta = Covariance (stock returns, market return) /Variance (market returns) 


Ab_beta=Ab_Data.cov().iloc[0][1]/Ab_Data.cov().iloc[0][0]
AC_beta=AC_Data.cov().iloc[0][1]/AC_Data.cov().iloc[0][0]
Co_beta=Co_Data.cov().iloc[0][1]/Co_Data.cov().iloc[0][0]
COS_beta=COS_Data.cov().iloc[0][1]/COS_Data.cov().iloc[0][0]
HCM_beta=HCM_Data.cov().iloc[0][1]/HCM_Data.cov().iloc[0][0]
HC_beta=HC_Data.cov().iloc[0][1]/HC_Data.cov().iloc[0][0]
IH_beta=IH_Data.cov().iloc[0][1]/IH_Data.cov().iloc[0][0]
Ja_beta=Ja_Data.cov().iloc[0][1]/Ja_Data.cov().iloc[0][0]
TO_beta=TO_Data.cov().iloc[0][1]/TO_Data.cov().iloc[0][0]
Ko_beta=Ko_Data.cov().iloc[0][1]/Ko_Data.cov().iloc[0][0]
LG_beta=LG_Data.cov().iloc[0][1]/LG_Data.cov().iloc[0][0]
Ne_beta=Ne_Data.cov().iloc[0][1]/Ne_Data.cov().iloc[0][0]


Cosmetics_betas = pd.DataFrame({
                        "AbleC&C": str(Ab_beta),
                        "Amorepacific": str(AC_beta),
                        "Cosmax": str(Co_beta),
                        "COSON": str(COS_beta),    
                        "Hankook Cosmetics Manufacturing": str(HCM_beta),
                        "Hankook Cosmetics": str(HC_beta),
                        "Its Hanbul": str(IH_beta),
                        "Jayjun Cosmetic": str(Ja_beta),
                        "TONYMOLY": str(TO_beta),
                        "Kolmar Korea": str(Ko_beta),
                        "LG Household": str(LG_beta),
                        "Korea Neopharm": str(Ne_beta),
},index=["Cosmetics Beta"])

Cosmetics_betas.head()

```

### Next we are going to observe the short-term impact of this event on the Cosmetic stock during our OBSERVATION WINDOW. Let's set a 30-day observation window from 2016-07-08 to 2017-08-08

### First, lets import the data after the event

```python
Ab_test=pdr.get_data_yahoo('078520.KS','2016-07-08','2016-08-08')['Adj Close']
AC_test=pdr.get_data_yahoo('090430.KS','2016-07-08','2016-08-08')['Adj Close']
Co_test=pdr.get_data_yahoo('192820.KS','2016-07-08','2016-08-08')['Adj Close']
```

```python
COS_test=pdr.get_data_yahoo('069110.KQ','2016-07-08','2016-08-08')['Adj Close']
HCM_test=pdr.get_data_yahoo('003350.KS','2016-07-08','2016-08-08')['Adj Close']
HC_test=pdr.get_data_yahoo('123690.KS','2016-07-08','2016-08-08')['Adj Close']
```

```python
IH_test=pdr.get_data_yahoo('226320.KS','2016-07-08','2016-08-08')['Adj Close']
Ja_test=pdr.get_data_yahoo('025620.KS','2016-07-08','2016-08-08')['Adj Close']
TO_test=pdr.get_data_yahoo('214420.KS','2016-07-08','2016-08-08')['Adj Close']
Ko_test=pdr.get_data_yahoo('024720.KS','2016-07-08','2016-08-08')['Adj Close']
```

```python
LG_test=pdr.get_data_yahoo('051900.KS','2016-07-08','2016-08-08')['Adj Close']
Ne_test=pdr.get_data_yahoo('092730.KQ','2016-07-08','2016-08-08')['Adj Close']
KS_test=pdr.get_data_yahoo('^KS11','2016-07-08','2016-08-08')['Adj Close']
```

```python
# Clean the data

Ab_test_ret=Ab_test.pct_change()
Ab_test_ret=pd.DataFrame(Ab_test_ret)
Ab_test_ret=Ab_test_ret.dropna() # here I dropped the NA terms

AC_test_ret=AC_test.pct_change()
AC_test_ret=pd.DataFrame(AC_test_ret)
AC_test_ret=AC_test_ret.dropna() # here I dropped the NA terms

Co_test_ret=Co_test.pct_change()
Co_test_ret=pd.DataFrame(Co_test_ret)
Co_test_ret=Co_test_ret.dropna() # here I dropped the NA terms

COS_test_ret=COS_test.pct_change()
COS_test_ret=pd.DataFrame(COS_test_ret)
COS_test_ret=COS_test_ret.dropna() # here I dropped the NA terms

HCM_test_ret=HCM_test.pct_change()
HCM_test_ret=pd.DataFrame(HCM_test_ret)
HCM_test_ret=HCM_test_ret.dropna() # here I dropped the NA terms

HC_test_ret=HC_test.pct_change()
HC_test_ret=pd.DataFrame(HC_test_ret)
HC_test_ret=HC_test_ret.dropna() # here I dropped the NA terms

```

```python
IH_test_ret=IH_test.pct_change()
IH_test_ret=pd.DataFrame(IH_test_ret)
IH_test_ret=IH_test_ret.dropna() # here I dropped the NA terms

Ja_test_ret=Ja_test.pct_change()
Ja_test_ret=pd.DataFrame(Ja_test_ret)
Ja_test_ret=Ja_test_ret.dropna() # here I dropped the NA terms

TO_test_ret=TO_test.pct_change()
TO_test_ret=pd.DataFrame(TO_test_ret)
TO_test_ret=TO_test_ret.dropna() # here I dropped the NA terms

Ko_test_ret=Ko_test.pct_change()
Ko_test_ret=pd.DataFrame(Ko_test_ret)
Ko_test_ret=Ko_test_ret.dropna() # here I dropped the NA terms

LG_test_ret=LG_test.pct_change()
LG_test_ret=pd.DataFrame(LG_test_ret)
LG_test_ret=LG_test_ret.dropna() # here I dropped the NA terms

Ne_test_ret=Ne_test.pct_change()
Ne_test_ret=pd.DataFrame(Ne_test_ret)
Ne_test_ret=Ne_test_ret.dropna() # here I dropped the NA terms

KS_test_ret=KS_test.pct_change()
KS_test_ret=pd.DataFrame(KS_test_ret)
KS_test_ret=KS_test_ret.dropna() # here I dropped the NA terms
```

```python
# First calculate abnormal returns
# Expected stock Aâ€™s return = Beta for stock A *(Market return - risk free rate) + risk free rate
# Abnormal Return = Actual return for the stock - Expected return for the stock

Ab_Abn_ret = Ab_test_ret-Ab_beta*(KS_test_ret-(0.015/365))+0.015/365
AC_Abn_ret = AC_test_ret-AC_beta*(KS_test_ret-(0.015/365))+0.015/365
Co_Abn_ret = Co_test_ret-Co_beta*(KS_test_ret-(0.015/365))+0.015/365
COS_Abn_ret = COS_test_ret-COS_beta*(KS_test_ret-(0.015/365))+0.015/365
HCM_Abn_ret = HCM_test_ret-HCM_beta*(KS_test_ret-(0.015/365))+0.015/365
HC_Abn_ret = HC_test_ret-HC_beta*(KS_test_ret-(0.015/365))+0.015/365
IH_Abn_ret = IH_test_ret-IH_beta*(KS_test_ret-(0.015/365))+0.015/365
Ja_Abn_ret = Ja_test_ret-Ja_beta*(KS_test_ret-(0.015/365))+0.015/365
TO_Abn_ret = TO_test_ret-TO_beta*(KS_test_ret-(0.015/365))+0.015/365
Ko_Abn_ret = Ko_test_ret-Ko_beta*(KS_test_ret-(0.015/365))+0.015/365
LG_Abn_ret = LG_test_ret-LG_beta*(KS_test_ret-(0.015/365))+0.015/365
Ne_Abn_ret = Ne_test_ret-Ne_beta*(KS_test_ret-(0.015/365))+0.015/365 
```

## THE IDEA FOR USING ABNORMAL RETURN IN EVENT STUDY IS: 
## THE CUMULATIVE ABNORMAL RETURN OF A STOCK AFTER CERTAIN EVENT HAPPENED IS CAUSED BY SUCH EVENT.

```python
# note here the import thing to measure impact is CUMULATIVE ABNORMAL RETURN, NOT ABNORMAL RETURN, next we calculate CUMULATIVE ABNORMAL RETURN
#Cumulative Abnormal Return is the sum of abnormal returns over a period (several days)

Ab_CAbn_ret=Ab_Abn_ret.cumsum()
AC_CAbn_ret=AC_Abn_ret.cumsum()
Co_CAbn_ret=Co_Abn_ret.cumsum()
COS_CAbn_ret=COS_Abn_ret.cumsum()
HCM_CAbn_ret=HCM_Abn_ret.cumsum()
HC_CAbn_ret=HC_Abn_ret.cumsum()
IH_CAbn_ret=IH_Abn_ret.cumsum()
Ja_CAbn_ret=Ja_Abn_ret.cumsum()
TO_CAbn_ret=TO_Abn_ret.cumsum()
Ko_CAbn_ret=Ko_Abn_ret.cumsum()
LG_CAbn_ret=LG_Abn_ret.cumsum()
Ne_CAbn_ret=Ne_Abn_ret.cumsum()

Cosmetics_CAbn_return = pd.DataFrame({
                        "AbleC&C": Ab_CAbn_ret["Adj Close"],
                        "Amorepacific": AC_CAbn_ret["Adj Close"],
                        "Cosmax": Co_CAbn_ret["Adj Close"],
                        "COSON": COS_CAbn_ret["Adj Close"],    
                        "Hankook Cosmetics Manufacturing": HCM_CAbn_ret["Adj Close"],
                        "Hankook Cosmetics": HC_CAbn_ret["Adj Close"],
                        "Its Hanbul": IH_CAbn_ret["Adj Close"],
                        "Jayjun Cosmetic": Ja_CAbn_ret["Adj Close"],
                        "TONYMOLY": TO_CAbn_ret["Adj Close"],
                        "Kolmar Korea": Ko_CAbn_ret["Adj Close"],
                        "LG Household": LG_CAbn_ret["Adj Close"],
                        "Korea Neopharm": Ne_CAbn_ret["Adj Close"],
})
 
Cosmetics_CAbn_return.tail(1)
```
