### In event studies, the beta is usually calculated using the data one year before the event. Next we are going to import the relevant data from yahoo and calculated beta

```python
AA_train=pdr.get_data_yahoo('020560.KS','2015-07-08','2016-07-08')['Adj Close']
HT_train=pdr.get_data_yahoo('039130.KS','2015-07-08','2016-07-08')['Adj Close']
Hj_train=pdr.get_data_yahoo('180640.KS','2015-07-08','2016-07-08')['Adj Close']

AA_train_ret=AA_train.pct_change()
AA_train_ret=pd.DataFrame(AA_train_ret)
AA_train_ret=AA_train_ret.dropna() # here I dropped the NA terms

HT_train_ret=HT_train.pct_change()
HT_train_ret=pd.DataFrame(HT_train_ret)
HT_train_ret=HT_train_ret.dropna() # here I dropped the NA terms

Hj_train_ret=Hj_train.pct_change()
Hj_train_ret=pd.DataFrame(Hj_train_ret)
Hj_train_ret=Hj_train_ret.dropna() # here I dropped the NA terms
```

```python
HS_train=pdr.get_data_yahoo('008770.KS','2015-07-08','2016-07-08')['Adj Close']
LTD_train=pdr.get_data_yahoo('032350.KS','2015-07-08','2016-07-08')['Adj Close']
MN_train=pdr.get_data_yahoo('080160.KQ','2015-07-08','2016-07-08')['Adj Close']

HS_train_ret=HS_train.pct_change()
HS_train_ret=pd.DataFrame(HS_train_ret)
HS_train_ret=HS_train_ret.dropna() # here I dropped the NA terms

LTD_train_ret=LTD_train.pct_change()
LTD_train_ret=pd.DataFrame(LTD_train_ret)
LTD_train_ret=LTD_train_ret.dropna() # here I dropped the NA terms

MN_train_ret=MN_train.pct_change()
MN_train_ret=pd.DataFrame(MN_train_ret)
MN_train_ret=MN_train_ret.dropna() # here I dropped the NA terms
```

```python
Pa_train=pdr.get_data_yahoo('034230.KQ','2015-07-08','2016-07-08')['Adj Close']
RT_train=pdr.get_data_yahoo('038390.KQ','2015-07-08','2016-07-08')['Adj Close']
Sj_train=pdr.get_data_yahoo('039310.KQ','2015-07-08','2016-07-08')['Adj Close']

Pa_train_ret=Pa_train.pct_change()
Pa_train_ret=pd.DataFrame(Pa_train_ret)
Pa_train_ret=Pa_train_ret.dropna() # here I dropped the NA terms

RT_train_ret=RT_train.pct_change()
RT_train_ret=pd.DataFrame(RT_train_ret)
RT_train_ret=RT_train_ret.dropna() # here I dropped the NA terms

Sj_train_ret=HT_train.pct_change()
Sj_train_ret=pd.DataFrame(Sj_train_ret)
Sj_train_ret=HT_train_ret.dropna() # here I dropped the NA terms
```

```python
SS_train=pdr.get_data_yahoo('004170.KS','2015-07-08','2016-07-08')['Adj Close']
VGT_train=pdr.get_data_yahoo('094850.KQ','2015-07-08','2016-07-08')['Adj Close']
KA_train=pdr.get_data_yahoo('003490.KS','2015-07-08','2016-07-08')['Adj Close']

SS_train_ret=SS_train.pct_change()
SS_train_ret=pd.DataFrame(SS_train_ret)
SS_train_ret=SS_train_ret.dropna() # here I dropped the NA terms

VGT_train_ret=VGT_train.pct_change()
VGT_train_ret=pd.DataFrame(VGT_train_ret)
VGT_train_ret=VGT_train_ret.dropna() # here I dropped the NA terms

KA_train_ret=KA_train.pct_change()
KA_train_ret=pd.DataFrame(KA_train_ret)
KA_train_ret=KA_train_ret.dropna() # here I dropped the NA terms
```

```python
KS_train=pdr.get_data_yahoo('^KS11','2015-07-08','2016-07-08')['Adj Close']

KS_train_ret=KS_train.pct_change()
KS_train_ret=pd.DataFrame(KS_train_ret)
KS_train_ret=KS_train_ret.dropna() # here I dropped the NA terms
```

```python
AA_Data=pd.merge(KS_train_ret,AA_train_ret,left_index=True,right_index=True)
HT_Data=pd.merge(KS_train_ret,HT_train_ret,left_index=True,right_index=True)
Hj_Data=pd.merge(KS_train_ret,Hj_train_ret,left_index=True,right_index=True) 
HS_Data=pd.merge(KS_train_ret,HS_train_ret,left_index=True,right_index=True) 
LTD_Data=pd.merge(KS_train_ret,LTD_train_ret,left_index=True,right_index=True) 
MN_Data=pd.merge(KS_train_ret,MN_train_ret,left_index=True,right_index=True) 
Pa_Data=pd.merge(KS_train_ret,Pa_train_ret,left_index=True,right_index=True) 
RT_Data=pd.merge(KS_train_ret,RT_train_ret,left_index=True,right_index=True) 
Sj_Data=pd.merge(KS_train_ret,Sj_train_ret,left_index=True,right_index=True) 
SS_Data=pd.merge(KS_train_ret,SS_train_ret,left_index=True,right_index=True) 
VGT_Data=pd.merge(KS_train_ret,VGT_train_ret,left_index=True,right_index=True) 
KA_Data=pd.merge(KS_train_ret,KA_train_ret,left_index=True,right_index=True) 


# It is usually safer to first merge two time-series using pd.merge(), it will make sure the dates between two series are matched

AA_Data.columns=['KS','AA']
HT_Data.columns=['KS','HT']
Hj_Data.columns=['KS','Hj']
HS_Data.columns=['KS','HS']
LTD_Data.columns=['KS','LTD']
MN_Data.columns=['KS','MN']
Pa_Data.columns=['KS','Pa']
RT_Data.columns=['KS','RT']
Sj_Data.columns=['KS','Sj']
SS_Data.columns=['KS','SS']
VGT_Data.columns=['KS','VGT']
KA_Data.columns=['KS','KA']
```

```python
# To calculate Beta
#Beta = Covariance (stock returns, market return) /Variance (market returns) 


AA_beta=AA_Data.cov().iloc[0][1]/AA_Data.cov().iloc[0][0]
HT_beta=HT_Data.cov().iloc[0][1]/HT_Data.cov().iloc[0][0]
Hj_beta=Hj_Data.cov().iloc[0][1]/Hj_Data.cov().iloc[0][0]
HS_beta=HS_Data.cov().iloc[0][1]/HS_Data.cov().iloc[0][0]
LTD_beta=LTD_Data.cov().iloc[0][1]/LTD_Data.cov().iloc[0][0]
MN_beta=MN_Data.cov().iloc[0][1]/MN_Data.cov().iloc[0][0]
Pa_beta=Pa_Data.cov().iloc[0][1]/Pa_Data.cov().iloc[0][0]
RT_beta=RT_Data.cov().iloc[0][1]/RT_Data.cov().iloc[0][0]
Sj_beta=Sj_Data.cov().iloc[0][1]/Sj_Data.cov().iloc[0][0]
SS_beta=SS_Data.cov().iloc[0][1]/SS_Data.cov().iloc[0][0]
VGT_beta=VGT_Data.cov().iloc[0][1]/VGT_Data.cov().iloc[0][0]
KA_beta=KA_Data.cov().iloc[0][1]/KA_Data.cov().iloc[0][0]


Tourism_betas = pd.DataFrame({
                        "Asiana": str(AA_beta),
                        "Hanatour": str(HT_beta),
                        "Hanjin": str(Hj_beta),
                        "Hotel Silla": str(HS_beta),    
                        "Lotte tour development": str(LTD_beta),
                        "Modotour network": str(MN_beta),
                        "Paradise": str(Pa_beta),
                        "Redcap Tour": str(RT_beta),
                        "Sejoong": str(Sj_beta),
                        "Sinsegae": str(SS_beta),
                        "Verygoodtour": str(VGT_beta),
                        "Korea airline": str(KA_beta),
},index=["Tourism Beta"])

Tourism_betas.head()

```

### First, lets import the data after the event

```python
AA_test=pdr.get_data_yahoo('020560.KS','2016-07-08','2016-08-08')['Adj Close']
HT_test=pdr.get_data_yahoo('039130.KS','2016-07-08','2016-08-08')['Adj Close']
Hj_test=pdr.get_data_yahoo('180640.KS','2016-07-08','2016-08-08')['Adj Close']
```

```python
HS_test=pdr.get_data_yahoo('008770.KS','2016-07-08','2016-08-08')['Adj Close']
LTD_test=pdr.get_data_yahoo('032350.KS','2016-07-08','2016-08-08')['Adj Close']
MN_test=pdr.get_data_yahoo('080160.KQ','2016-07-08','2016-08-08')['Adj Close']
```

```python
Pa_test=pdr.get_data_yahoo('034230.KQ','2016-07-08','2016-08-08')['Adj Close']
RT_test=pdr.get_data_yahoo('038390.KQ','2016-07-08','2016-08-08')['Adj Close']
Sj_test=pdr.get_data_yahoo('039310.KQ','2016-07-08','2016-08-08')['Adj Close']
SS_test=pdr.get_data_yahoo('004170.KS','2016-07-08','2016-08-08')['Adj Close']
```

```python
VGT_test=pdr.get_data_yahoo('094850.KQ','2016-07-08','2016-08-08')['Adj Close']
KA_test=pdr.get_data_yahoo('003490.KS','2016-07-08','2016-08-08')['Adj Close']
KS_test=pdr.get_data_yahoo('^KS11','2016-07-08','2016-08-08')['Adj Close']
```

```python
# Clean the data

AA_test_ret=AA_test.pct_change()
AA_test_ret=pd.DataFrame(AA_test_ret)
AA_test_ret=AA_test_ret.dropna() # here I dropped the NA terms

HT_test_ret=HT_test.pct_change()
HT_test_ret=pd.DataFrame(HT_test_ret)
HT_test_ret=HT_test_ret.dropna() # here I dropped the NA terms

Hj_test_ret=Hj_test.pct_change()
Hj_test_ret=pd.DataFrame(Hj_test_ret)
Hj_test_ret=Hj_test_ret.dropna() # here I dropped the NA terms

HS_test_ret=HS_test.pct_change()
HS_test_ret=pd.DataFrame(HS_test_ret)
HS_test_ret=HS_test_ret.dropna() # here I dropped the NA terms

LTD_test_ret=LTD_test.pct_change()
LTD_test_ret=pd.DataFrame(LTD_test_ret)
LTD_test_ret=LTD_test_ret.dropna() # here I dropped the NA terms

MN_test_ret=MN_test.pct_change()
MN_test_ret=pd.DataFrame(MN_test_ret)
MN_test_ret=MN_test_ret.dropna() # here I dropped the NA terms

```

```python
Pa_test_ret=Pa_test.pct_change()
Pa_test_ret=pd.DataFrame(Pa_test_ret)
Pa_test_ret=Pa_test_ret.dropna() # here I dropped the NA terms

RT_test_ret=RT_test.pct_change()
RT_test_ret=pd.DataFrame(RT_test_ret)
RT_test_ret=RT_test_ret.dropna() # here I dropped the NA terms

Sj_test_ret=HT_test.pct_change()
Sj_test_ret=pd.DataFrame(Sj_test_ret)
Sj_test_ret=HT_test_ret.dropna() # here I dropped the NA terms

SS_test_ret=SS_test.pct_change()
SS_test_ret=pd.DataFrame(SS_test_ret)
SS_test_ret=SS_test_ret.dropna() # here I dropped the NA terms

VGT_test_ret=VGT_test.pct_change()
VGT_test_ret=pd.DataFrame(VGT_test_ret)
VGT_test_ret=VGT_test_ret.dropna() # here I dropped the NA terms

KA_test_ret=KA_test.pct_change()
KA_test_ret=pd.DataFrame(KA_test_ret)
KA_test_ret=KA_test_ret.dropna() # here I dropped the NA terms

KS_test_ret=KS_test.pct_change()
KS_test_ret=pd.DataFrame(KS_test_ret)
KS_test_ret=KS_test_ret.dropna() # here I dropped the NA terms
```

```python
# First calculate abnormal returns
# Expected stock Aâ€™s return = Beta for stock A *(Market return - risk free rate) + risk free rate
# Abnormal Return = Actual return for the stock - Expected return for the stock

AA_Abn_ret = AA_test_ret-AA_beta*(KS_test_ret-(0.015/365))+0.015/365
HT_Abn_ret = HT_test_ret-HT_beta*(KS_test_ret-(0.015/365))+0.015/365
Hj_Abn_ret = Hj_test_ret-Hj_beta*(KS_test_ret-(0.015/365))+0.015/365
HS_Abn_ret = HS_test_ret-HS_beta*(KS_test_ret-(0.015/365))+0.015/365
LTD_Abn_ret = LTD_test_ret-LTD_beta*(KS_test_ret-(0.015/365))+0.015/365
MN_Abn_ret = MN_test_ret-MN_beta*(KS_test_ret-(0.015/365))+0.015/365
Pa_Abn_ret = Pa_test_ret-Pa_beta*(KS_test_ret-(0.015/365))+0.015/365
RT_Abn_ret = RT_test_ret-RT_beta*(KS_test_ret-(0.015/365))+0.015/365
Sj_Abn_ret = Sj_test_ret-Sj_beta*(KS_test_ret-(0.015/365))+0.015/365
SS_Abn_ret = SS_test_ret-SS_beta*(KS_test_ret-(0.015/365))+0.015/365
VGT_Abn_ret = VGT_test_ret-VGT_beta*(KS_test_ret-(0.015/365))+0.015/365
KA_Abn_ret = KA_test_ret-KA_beta*(KS_test_ret-(0.015/365))+0.015/365 
```

## THE IDEA FOR USING ABNORMAL RETURN IN EVENT STUDY IS: 
## THE CUMULATIVE ABNORMAL RETURN OF A STOCK AFTER CERTAIN EVENT HAPPENED IS CAUSED BY SUCH EVENT.

```python
# note here the import thing to measure impact is CUMULATIVE ABNORMAL RETURN, NOT ABNORMAL RETURN, next we calculate CUMULATIVE ABNORMAL RETURN
#Cumulative Abnormal Return is the sum of abnormal returns over a period (several days)

AA_CAbn_ret=AA_Abn_ret.cumsum()
HT_CAbn_ret=HT_Abn_ret.cumsum()
Hj_CAbn_ret=Hj_Abn_ret.cumsum()
HS_CAbn_ret=HS_Abn_ret.cumsum()
LTD_CAbn_ret=LTD_Abn_ret.cumsum()
MN_CAbn_ret=MN_Abn_ret.cumsum()
Pa_CAbn_ret=Pa_Abn_ret.cumsum()
RT_CAbn_ret=RT_Abn_ret.cumsum()
Sj_CAbn_ret=Sj_Abn_ret.cumsum()
SS_CAbn_ret=SS_Abn_ret.cumsum()
VGT_CAbn_ret=VGT_Abn_ret.cumsum()
KA_CAbn_ret=KA_Abn_ret.cumsum()

Tourism_CAbn_return = pd.DataFrame({
                        "Asiana": AA_CAbn_ret["Adj Close"],
                        "Hanatour": HT_CAbn_ret["Adj Close"],
                        "Hanjin": Hj_CAbn_ret["Adj Close"],
                        "Hotel Silla": HS_CAbn_ret["Adj Close"],    
                        "Lotte tour development": LTD_CAbn_ret["Adj Close"],
                        "Modotour network": MN_CAbn_ret["Adj Close"],
                        "Paradise": Pa_CAbn_ret["Adj Close"],
                        "Redcap Tour": RT_CAbn_ret["Adj Close"],
                        "Sejoong": Sj_CAbn_ret["Adj Close"],
                        "Sinsegae": SS_CAbn_ret["Adj Close"],
                        "Verygoodtour": VGT_CAbn_ret["Adj Close"],
                        "Korea airline": KA_CAbn_ret["Adj Close"],
})
 
Tourism_CAbn_return.tail(1)
```
